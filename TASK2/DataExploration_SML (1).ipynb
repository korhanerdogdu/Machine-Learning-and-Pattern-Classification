```python
import numpy as np
import pandas as pd
from scipy.stats import skew, kurtosis

# Load the numpy array
development_data = np.load('/Users/korhanerdogdu/Desktop/development.npy')

# Initialize a list to hold the distribution statistics for each feature
features_stats = []

# Calculate stats for each feature
for i in range(development_data.shape[1]):  # Assuming the second dimension is features
    feature_data = development_data[:, i, :].flatten()
    mean = np.mean(feature_data)
    variance = np.var(feature_data)
    skewness = skew(feature_data)
    kurt = kurtosis(feature_data)
    
    # Append the stats for this feature to the list
    features_stats.append([mean, variance, skewness, kurt])

# Convert the list to a DataFrame for easier analysis
features_df = pd.DataFrame(features_stats, columns=['Mean', 'Variance', 'Skewness', 'Kurtosis'])

# Compute the correlation matrix for the statistical measures
correlation_matrix = features_df.corr()

# Plot the correlation matrix
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Matrix of Distribution Statistics')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import skew, kurtosis

def feature_correlations(data, feature_indices):
    """
    Calculate and plot the correlation matrix for a selection of features based on their distributions.
    
    Parameters:
    - data: The full dataset as a numpy array with dimensions (snippets, features, frames).
    - feature_indices: A list of indices for the features to be analyzed.
    """
    # Initialize a dictionary to hold our feature statistics
    feature_stats = {'Mean': [], 'StdDev': [], 'Skewness': [], 'Kurtosis': []}
    
    # Calculate statistics for each selected feature
    for idx in feature_indices:
        feature_data = data[:, idx, :].flatten()  # Flatten the data for this feature
        feature_stats['Mean'].append(np.mean(feature_data))
        feature_stats['StdDev'].append(np.std(feature_data))
        feature_stats['Skewness'].append(skew(feature_data))
        feature_stats['Kurtosis'].append(kurtosis(feature_data))
    
    # Convert the dictionary to a DataFrame for easier correlation calculation
    import pandas as pd
    stats_df = pd.DataFrame(feature_stats)
    
    # Calculate the correlation matrix
    correlation_matrix = stats_df.corr()
    
    # Plot the correlation matrix
    plt.figure(figsize=(8, 6))
    sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', cbar=True)
    plt.title('Correlation Matrix Among Feature Statistics')
    plt.show()

# Example usage with the first 10 features
feature_indices = list(range(10))  # Selecting the first 10 features
feature_correlations(development_data, feature_indices)
```

```python
import numpy as np
import pandas as pd

def create_feature_means_dataframe(data, feature_names):
    """
    For each sample in the dataset, calculate the mean of each feature over time,
    and compile these results into a DataFrame.

    Parameters:
    - data: The dataset containing all samples and features, with shape (samples, features, timestamps).
    - feature_names: A list of all feature names in the dataset.
    
    Returns:
    - DataFrame with each row representing a sample and each column a feature mean over time.
    """
    # Initialize an empty list to store the mean values for each sample
    samples_feature_means = []
    
    # Loop through each sample
    for sample_index in range(data.shape[0]):
        # Calculate the mean of each feature over time for the current sample
        sample_means = data[sample_index, :, :].mean(axis=1)
        samples_feature_means.append(sample_means)
    
    # Create a DataFrame from the list of mean values
    feature_means_df = pd.DataFrame(samples_feature_means, columns=feature_names)
    
    return feature_means_df

# Example usage
# Assuming 'data' is your loaded dataset with shape (samples, features, timestamps)
# and 'feature_names' contains the names of all 175 features in the correct order
#feature_names = [f'Feature_{i+1}' for i in range(175)]  # Example feature names

feature_means_df = create_feature_means_dataframe(data, feature_names)
print(feature_means_df.head(30))  # Print the first few rows of the new DataFrame
```

```python
import numpy as np
import pandas as pd

def create_combined_stats_dataframe(data, feature_names, sample_ids):
    """
    Calculate mean, median, and standard deviation for each feature across all samples,
    compiling these results into a DataFrame that includes sample identifiers.

    Parameters:
    - data: The dataset containing all samples and features, shape (samples, features, timestamps).
    - feature_names: A list of all feature names in the dataset.
    - sample_ids: A list or array of identifiers for each sample.
    
    Returns:
    - A DataFrame where each row represents a sample and includes the sample identifier,
      with columns for the calculated statistics for all features.
    """
    # Initialize a list to store data for DataFrame creation
    stats_data = []
    
    # Calculate the statistics for each sample
    for i in range(data.shape[0]):
        sample_means = np.mean(data[i, :, :], axis=1)
        sample_medians = np.median(data[i, :, :], axis=1)
        sample_stddevs = np.std(data[i, :, :], axis=1)
        
        # Create a dictionary for the current sample's statistics
        sample_stats = {'Sample_ID': sample_ids[i]}
        for j, feature_name in enumerate(feature_names):
            sample_stats[f'{feature_name}_Mean'] = sample_means[j]
            sample_stats[f'{feature_name}_Median'] = sample_medians[j]
            sample_stats[f'{feature_name}_StdDev'] = sample_stddevs[j]
        
        # Add the sample's statistics to the list
        stats_data.append(sample_stats)
    
    # Convert the list of dictionaries to a DataFrame
    stats_df = pd.DataFrame(stats_data)
    
    return stats_df

# Example usage:
# Assuming 'data' is your dataset with the shape (samples, features, timestamps)
sample_ids = [f'Sample_{i+1}' for i in range(data.shape[0])]  # Generating sample identifiers

combined_stats_df = create_combined_stats_dataframe(data, feature_names, sample_ids)
print(combined_stats_df.head())  # Display the first few rows to check the DataFrame
```

```python
import numpy as np
import pandas as pd

def transform_data_to_dataframe(data, feature_names):
    """
    Transform the dataset from a 3D array into a 2D DataFrame with each row as a sample
    and each column as a flattened feature across all timestamps.

    Parameters:
    - data: The dataset containing all samples, features, and timestamps, 
            with shape (samples, features, timestamps).
    - feature_names: A list of original feature names before flattening.
    
    Returns:
    - A DataFrame with samples as rows and 7700 flattened features as columns.
    """
    # Calculate the total number of features after flattening
    num_samples, num_features, num_timestamps = data.shape
    total_features = num_features * num_timestamps
    
    # Flatten the features and timestamps dimensions
    flattened_data = data.reshape(num_samples, total_features)
    
    # Generate new column names for each flattened feature
    flattened_feature_names = [f'{feature_name}_T{timestamp+1}' 
                               for feature_name in feature_names 
                               for timestamp in range(num_timestamps)]
    
    # Create and return the DataFrame
    flattened_df = pd.DataFrame(flattened_data, columns=flattened_feature_names)
    
    return flattened_df


# Assuming 'data' is your dataset with shape (samples, 175 features, 44 timestamps)
flattened_df = transform_data_to_dataframe(data, feature_names)
print(flattened_df.head())  # Display the first few rows of the new DataFrame
```

***Central Tendency:***

The median value for each word is indicated by the line within each box. It seems the median values for "Haus" and "aus" are somewhat similar across the timestamps, which might suggest a degree of similarity in their feature distributions.

***Spread of Data:***

The height of the boxes (the IQR) shows the middle 50% of the data. A taller box indicates more variability within the interquartile range. For instance, "Haus" and "aus" show a similar spread across all features, suggesting that the variability in these features is somewhat consistent for these two words.

***Outliers:***

The individual points beyond the whiskers are outliers. Many outliers suggest there are samples for each word where the feature values are significantly different from the median.

***Comparison Across Words:***

Comparing similar words (e.g., "Haus" and "aus", "Ofen" and "offen", "nicht" and "Licht"), the distributions for "Haus" vs. "aus" and "Ofen" vs. "offen" show some similarities in their median and spread, especially when compared to the distributions between "nicht" and "Licht". However, there are notable differences in the median and IQR among these pairs, indicating they have distinctions in their feature values which could be useful for classification.

***Timestamp Differences:***

There does not seem to be a drastic difference in the distributions of the features across different timestamps, suggesting that the bandwidth measurements are somewhat stable over the four timestamps represented here.

***Comparative Analysis:***

For classification tasks, features with overlapping distributions might not be as discriminative on their own. If "Haus" and "aus" consistently show similar distributions across various features, they might be harder to differentiate based solely on those features.
